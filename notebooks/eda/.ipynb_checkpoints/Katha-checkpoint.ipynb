```python
import pandas as pd
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
from wordcloud import WordCloud
from stop_words import stop_bag
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

# 1. Import raw data and get an overview

```python
df = pd.read_csv("comm.csv")

df = df.rename({"NPS Kommentar TP/CJ":"comment"}, axis=1)
df.head()
```

0-6 neg
7-10 pos

```python
df.groupby("NPS").count()
```

```python
df_neg = df.loc[(df.NPS <= 6)]
df_neg.groupby("NPS").count()
df_neg.describe().T
```

```python
df_pos = df.loc[(df.NPS > 6)]
df_pos.groupby("NPS").count()
df_pos.describe().T
```

```python
df_neg_comm = df_neg.comment
df_neg_comm = df_neg_comm.dropna()
df_neg_comm.isna().sum()
comments_neg = list(df_neg_comm)
type(comments_neg)
```

```python
df_pos_comm = df_pos.comment
df_pos_comm = df_pos_comm.dropna()
df_pos_comm.isna().sum()
comments_pos = list(df_pos_comm)
type(comments_pos)
```

```python
x, y = np.ogrid[:300, :300]

mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)
```

```python
comments = comments_neg

if comments == comments_neg:
    title="Negative Kommentare"
else:
    title="Positive Kommentare"
```

```python
stop_words = stop_bag.split("\n")

word_list = ",".join(comments)
word_list = word_list.lower().split(" ")
comments_cleaned = [word for word in word_list if word not in stop_words]
comments_cleaned = ",".join(comments_cleaned)

wordcloud = WordCloud(background_color='white', width = 500, height = 500, mask=mask, max_words=100)\
                    .generate(comments_cleaned)

plt.axis("off")
#plt.margins(x=1, y=1)
plt.rcParams["figure.figsize"] = (800,800)
plt.imshow(wordcloud)
plt.title(title)
plt.show()
plt.close()
```
